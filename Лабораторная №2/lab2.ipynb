```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
```

```python
pd.read_csv('nfcapd.csv')
```

```python
data = pd.read_csv('nfcapd.csv')[:17448]
```

```python
data.ts = pd.to_datetime(data.ts)
data.da = data.da.astype(str)
```

```python
l = data.loc[(data['sa'] == '87.245.198.147') | (data['da'] == '87.245.198.147')].ibyt.tolist()
for i in range(1,len(l)):
    l[i] += l[i-1]
    
```

```python
def internet(ip, k):
    X = data.loc[(data['sa'] == ip) | (data['da'] == ip)].ibyt.sum() * k
    plt.plot(data.loc[(data['sa'] == ip) | (data['da'] == ip)].ts, l)
    plt.title("Traffic(time)")
    plt.xlabel("time")
    plt.ylabel("volume of traffic")
    return(X)
```

# Ответ
